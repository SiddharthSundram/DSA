Given n, print a solid square of * of size n.

```python
def solidSqr(n):
    for i in range(n):
        for j in range(n):
            print("* ", end=' ')
        print()

n = int(input())
solidSqr(n)
```

Pattern 2 – Right-Angled Triangle (Stars)

```python
def rightPattern(n):
    for i in range(n):
        for j in range(n):
            if i >= j:
                print("* ", end=' ')
            else:
                break
        print()
        
n = int(input())
rightPattern(n)
```

Pattern 3 – Right-Angled Triangle (Numbers)

```python
def  rightNum(n):
    for i in range(n):
        for j in range(n):
            if i >= j:
                print(j+1,end=' ')
            else:
                break
        print()
        
n = int(input())
rightNum(n)
```
