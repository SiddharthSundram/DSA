Given a string name and an integer n, print the name n times.

```python
def nameTimes(name,time):
    for i in range(time):
        print(name)

name = input()
t = int(input())
nameTimes(name,t)
```

Print all numbers from 1 to n.

```python
def printall(n):
    for i in range(1,n+1):
        print(i)

i = int(input())
printall(i)
```

Print all numbers from n down to 1.

```python
def printBack(n):
    # while n >0:
    #     print(n)
    #     n -= 1
    
    for i in range(n,0,-1):
        print(i)

n = int(input())
printBack(n)
```

Find the sum of the first n natural numbers.

```python
def addi(n):
    # if n >= 1:
    #     n = int((n* (n+1))/2)
    #     print(n)
    sum = 0
    for i in range(n+1):
        sum += i
    print(sum)

n = int(input())
addi(n)
    
```

Check Positive, Negative, or Zero

```python
def checkNum(n):
    if n > 0:
        print("Positive")
    elif n == 0:
        print('Zero')
    else:
        print("Negative")
n = int(input())
checkNum(n)
```

Given two integers a and b, print the larger one.
If equal, print "Equal".

```python
def compareNum(a,b):
    if a > b:
        print(a)
    elif b>a :
        print(b)
    else:
        print('Equal')
        
a,b = map(int, input().split())
compareNum(a,b)
```

Given an array, reverse it.

```python
def revArr(n,num):
    for i in range(n-1,-1,-1):
        print(num[i],end=' ')
    
n = int(input())
num = list(map(int,input().split()))
revArr(n,num)
```
