```python
def printall(n):
    for i in range(1,n+1):
        print(i, end=' ')

i = int(input())
printall(i)
```

Print the sum of numbers from L to R.

```python
def sumn(l,r):
    total= 0
    for i in range(l,r+1):
        total += i
    return total

a,b = map(int,input().split())
sumn(a,b)
```

```python
import math

def checkp(n):
    if n <= 1:
        print("Not Prime")
        return
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            print("Not Prime")
            return
    print("Prime")

n = int(input())
checkp(n)
```

Given an array, reverse it.

```python
def revArr(n,num):
    for i in range(n-1,-1,-1):
        print(num[i],end=' ')
    
n = int(input())
num = list(map(int,input().split()))
revArr(n,num)
```

Check if an array is sorted in non-decreasing order.

```python
def checkDec(n,arr):
    for i in range(1,n):
        if arr[i] < arr[i-1]:
            return False
        return True
            
n = int(input())
arr = list(map(int,input().split()))
out =checkDec(n,arr)
if out == True:
    print("YES")
else:
    print("NO")
```
